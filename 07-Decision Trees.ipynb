```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from matplotlib import pyplot as plt
X,y = make_classification(n_samples=100,n_features=5,n_informative=2,n_redundant=1,random_state=1)
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,test_size=0.2)
dtree= DecisionTreeClassifier(random_state=0,max_depth=5)
dtree.fit(Xtrain,Ytrain)
r=export_text(dtree)
print(r)
```

```python
fig=plt.figure(figsize=(10,15))
_=tree.plot_tree(dtree,filled=True,fontsize=12)
```

```python
Ypred=dtree.predict(Xtest)
print(Ytest,"\n",Ypred)
```

```python
cf = confusion_matrix(Ytest,Ypred)
print(cf)
accuracy = (cf[0][0]+cf[1][1])/(cf[0][0]+cf[0][1]+cf[1][0]+cf[1][1])
print(accuracy*100)
```
