```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_text
from matplotlib import pyplot as plt
X,y = make_classification(n_samples=100,n_features=5,n_informative=2,n_redundant=1,random_state=1)
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,test_size=0.2)
rtree= RandomForestClassifier(n_estimators=50,random_state=15,max_depth=5)
rtree.fit(Xtrain,Ytrain)
```

```python
r=export_text(rtree.estimators_[0])
print(r)
fig=plt.figure(figsize=(10,15))
#For Base Estimator
_=tree.plot_tree(rtree.estimators_[0],filled=True,fontsize=12)
```

```python
Ypred=rtree.predict(Xtest)
print(Ytest,"\n",Ypred)
```

```python
cf = confusion_matrix(Ytest,Ypred)
print(cf)
accuracy = (cf[0][0]+cf[1][1])/(cf[0][0]+cf[0][1]+cf[1][0]+cf[1][1])
print(accuracy*100)
```
