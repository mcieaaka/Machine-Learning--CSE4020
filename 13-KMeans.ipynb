```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from sklearn.metrics import confusion_matrix
X,y = make_blobs(n_samples=1000,n_features=3,centers=3,random_state=0)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(X)
import matplotlib.pyplot as plt

filtered_label0 = X[kmeans == 0]
filtered_label1 = X[kmeans == 1]
filtered_label2 = X[kmeans == 2]
 
#plotting the results
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1],color='blue')
plt.scatter(filtered_label1[:,0] , filtered_label1[:,1],color='red')
plt.scatter(filtered_label2[:,0] , filtered_label2[:,1],color='green')
plt.show()
```
